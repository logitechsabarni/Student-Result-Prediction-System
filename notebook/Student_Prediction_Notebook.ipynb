```python
import pandas as pd
import numpy as np

np.random.seed(42)

num_rows = 3000

# ----------------------------
# Academic Core Metrics
# ----------------------------

attendance = np.clip(np.random.normal(75, 15, num_rows), 40, 100)
internal1 = np.clip(attendance * 0.6 + np.random.normal(10, 8, num_rows), 30, 100)
internal2 = np.clip(attendance * 0.65 + np.random.normal(8, 10, num_rows), 30, 100)
assignment_avg = np.clip(attendance * 0.7 + np.random.normal(5, 7, num_rows), 40, 100)
quiz_avg = np.clip(internal1 * 0.6 + np.random.normal(5, 8, num_rows), 30, 100)
lab_perf = np.clip(assignment_avg * 0.8 + np.random.normal(5, 6, num_rows), 40, 100)
previous_gpa = np.clip((internal1 + internal2)/20 + np.random.normal(0, 0.5, num_rows), 4, 10)
backlogs = np.random.poisson(0.5, num_rows)

project_score = np.clip(assignment_avg * 0.8 + np.random.normal(5, 7, num_rows), 40, 100)
midterm_score = np.clip((internal1 + internal2)/2 + np.random.normal(5, 6, num_rows), 40, 100)

# ----------------------------
# Behavioral Features
# ----------------------------

study_hours = np.clip(np.random.normal(3, 1.5, num_rows), 0.5, 8)
library_visits = np.random.poisson(5, num_rows)
online_courses = np.random.poisson(2, num_rows)
doubt_sessions = np.random.poisson(3, num_rows)

# ----------------------------
# Demographics
# ----------------------------

age = np.random.randint(18, 25, num_rows)
family_income = np.clip(np.random.normal(500000, 200000, num_rows), 100000, 1500000)
distance = np.clip(np.random.normal(10, 5, num_rows), 1, 40)

gender = np.random.choice(["Male", "Female"], num_rows)
parent_education = np.random.choice(
    ["School", "Graduate", "Postgraduate"], num_rows
)
urban_rural = np.random.choice(["Urban", "Rural"], num_rows)

# ----------------------------
# Lifestyle
# ----------------------------

sleep_hours = np.clip(np.random.normal(6.5, 1.5, num_rows), 4, 9)
stress_level = np.clip(np.random.normal(5, 2, num_rows), 1, 10)
social_media = np.clip(np.random.normal(3, 1.5, num_rows), 0.5, 8)

internet_access = np.random.choice(["Yes", "No"], num_rows)
participation_club = np.random.choice(["Yes", "No"], num_rows)
participation_sports = np.random.choice(["Yes", "No"], num_rows)
internship_completed = np.random.choice(["Yes", "No"], num_rows)
health_issues = np.random.choice(["Yes", "No"], num_rows)
part_time_job = np.random.choice(["Yes", "No"], num_rows)

# ----------------------------
# Final Score (Target)
# Strong Correlation Formula
# ----------------------------

final_exam_score = (
    attendance * 0.15 +
    internal1 * 0.2 +
    internal2 * 0.2 +
    assignment_avg * 0.15 +
    study_hours * 2 +
    previous_gpa * 5 -
    stress_level * 1.5 -
    backlogs * 5 +
    np.random.normal(0, 5, num_rows)
)

final_exam_score = np.clip(final_exam_score, 30, 100)

# Pass/Fail Classification
pass_fail = np.where(final_exam_score >= 50, "Pass", "Fail")

# ----------------------------
# Create DataFrame
# ----------------------------

data = pd.DataFrame({
    "attendance_percentage": attendance,
    "internal_exam_1": internal1,
    "internal_exam_2": internal2,
    "assignment_score_avg": assignment_avg,
    "quiz_avg": quiz_avg,
    "lab_performance": lab_perf,
    "previous_sem_gpa": previous_gpa,
    "backlog_count": backlogs,
    "project_score": project_score,
    "midterm_score": midterm_score,
    "study_hours_per_day": study_hours,
    "library_visits_per_month": library_visits,
    "online_course_count": online_courses,
    "doubt_sessions_attended": doubt_sessions,
    "age": age,
    "family_income": family_income,
    "distance_from_college_km": distance,
    "gender": gender,
    "parental_education_level": parent_education,
    "urban_or_rural": urban_rural,
    "sleep_hours": sleep_hours,
    "stress_level": stress_level,
    "social_media_hours": social_media,
    "internet_access_home": internet_access,
    "participation_club": participation_club,
    "participation_sports": participation_sports,
    "internship_completed": internship_completed,
    "health_issues": health_issues,
    "part_time_job": part_time_job,
    "final_exam_score": final_exam_score,
    "pass_fail": pass_fail
})

data.to_csv("student_performance_advanced.csv", index=False)

print("Dataset generated successfully with 3000 rows.")
print(data.head())
```

```python
import pandas as pd

data = pd.read_csv("student_performance_advanced.csv")
data.head()
```

```python
# Regression Target
y_reg = data["final_exam_score"]

# Classification Target
y_clf = data["pass_fail"]

# Drop targets from features
X = data.drop(columns=["final_exam_score", "pass_fail"])
```

```python
numerical_features = X.select_dtypes(include=["int64", "float64"]).columns
categorical_features = X.select_dtypes(include=["object"]).columns

print("Numerical:", len(numerical_features))
print("Categorical:", len(categorical_features))
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import train_test_split

numeric_pipeline = Pipeline([
    ("scaler", StandardScaler())
])

categorical_pipeline = Pipeline([
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer([
    ("num", numeric_pipeline, numerical_features),
    ("cat", categorical_pipeline, categorical_features)
])
```

```python
reg_pipeline = Pipeline([
    ("preprocessing", preprocessor),
    ("feature_selection", SelectKBest(score_func=f_regression, k=20)),
    ("model", RandomForestRegressor(random_state=42))
])

X_train, X_test, y_train, y_test = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

reg_pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

y_pred = reg_pipeline.predict(X_test)

print("R2 Score:", r2_score(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
clf_pipeline = Pipeline([
    ("preprocessing", preprocessor),
    ("model", RandomForestClassifier(random_state=42))
])

X_train, X_test, y_train, y_test = train_test_split(
    X, y_clf, test_size=0.2, random_state=42
)

clf_pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = clf_pipeline.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
import shap

# Fit only preprocessing
X_processed = reg_pipeline.named_steps["preprocessing"].fit_transform(X)

# Train a fresh RF for SHAP (needed for tree explainer)
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_processed, y_reg)

explainer = shap.TreeExplainer(rf_model)
shap_values = explainer.shap_values(X_processed[:200])  # limit for speed

shap.summary_plot(shap_values, X_processed[:200])
```

```python
import joblib

joblib.dump(reg_pipeline, "regression_model.pkl")
joblib.dump(clf_pipeline, "classification_model.pkl")
```

```python
import pandas as pd
import numpy as np
import joblib
import shap
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import r2_score, mean_absolute_error, accuracy_score, classification_report

import sklearn
print("Sklearn version:", sklearn.__version__)
```

```python
df = pd.read_csv("student_performance_advanced.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())

plt.figure(figsize=(10,5))
sns.histplot(df["final_exam_score"], bins=30)
plt.title("Final Exam Score Distribution")
plt.show()
```

```python
X = df.drop(["final_exam_score", "pass_fail"], axis=1)
y_reg = df["final_exam_score"]
y_clf = df["pass_fail"]

numeric_features = X.select_dtypes(include=["int64", "float64"]).columns
categorical_features = X.select_dtypes(include=["object"]).columns

print("Numeric features:", len(numeric_features))
print("Categorical features:", len(categorical_features))
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ]
)
```

```python
reg_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", RandomForestRegressor(n_estimators=300, random_state=42))
])

clf_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(n_estimators=300, random_state=42))
])
```

```python
X_train, X_test, y_reg_train, y_reg_test = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

_, _, y_clf_train, y_clf_test = train_test_split(
    X, y_clf, test_size=0.2, random_state=42
)
```

```python
reg_pipeline.fit(X_train, y_reg_train)
clf_pipeline.fit(X_train, y_clf_train)

print("Training Complete ✅")
```

```python
reg_pred = reg_pipeline.predict(X_test)

print("R2 Score:", r2_score(y_reg_test, reg_pred))
print("MAE:", mean_absolute_error(y_reg_test, reg_pred))
```

```python
clf_pred = clf_pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_clf_test, clf_pred))
print(classification_report(y_clf_test, clf_pred))
```

```python
explainer = shap.TreeExplainer(reg_pipeline.named_steps["model"])

X_sample = reg_pipeline.named_steps["preprocessor"].transform(X_test[:100])

shap_values = explainer.shap_values(X_sample)

shap.summary_plot(shap_values, X_sample)
```

```python
joblib.dump(reg_pipeline, "regression_model.pkl")
joblib.dump(clf_pipeline, "classification_model.pkl")

print("Models saved successfully!")
```
