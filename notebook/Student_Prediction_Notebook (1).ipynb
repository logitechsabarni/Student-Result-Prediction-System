```python
import pandas as pd
import numpy as np

np.random.seed(42)

num_rows = 3000

# ----------------------------
# Academic Core Metrics
# ----------------------------

attendance = np.clip(np.random.normal(75, 15, num_rows), 40, 100)
internal1 = np.clip(attendance * 0.6 + np.random.normal(10, 8, num_rows), 30, 100)
internal2 = np.clip(attendance * 0.65 + np.random.normal(8, 10, num_rows), 30, 100)
assignment_avg = np.clip(attendance * 0.7 + np.random.normal(5, 7, num_rows), 40, 100)
quiz_avg = np.clip(internal1 * 0.6 + np.random.normal(5, 8, num_rows), 30, 100)
lab_perf = np.clip(assignment_avg * 0.8 + np.random.normal(5, 6, num_rows), 40, 100)
previous_gpa = np.clip((internal1 + internal2)/20 + np.random.normal(0, 0.5, num_rows), 4, 10)
backlogs = np.random.poisson(0.5, num_rows)

project_score = np.clip(assignment_avg * 0.8 + np.random.normal(5, 7, num_rows), 40, 100)
midterm_score = np.clip((internal1 + internal2)/2 + np.random.normal(5, 6, num_rows), 40, 100)

# ----------------------------
# Behavioral Features
# ----------------------------

study_hours = np.clip(np.random.normal(3, 1.5, num_rows), 0.5, 8)
library_visits = np.random.poisson(5, num_rows)
online_courses = np.random.poisson(2, num_rows)
doubt_sessions = np.random.poisson(3, num_rows)

# ----------------------------
# Demographics
# ----------------------------

age = np.random.randint(18, 25, num_rows)
family_income = np.clip(np.random.normal(500000, 200000, num_rows), 100000, 1500000)
distance = np.clip(np.random.normal(10, 5, num_rows), 1, 40)

gender = np.random.choice(["Male", "Female"], num_rows)
parent_education = np.random.choice(
    ["School", "Graduate", "Postgraduate"], num_rows
)
urban_rural = np.random.choice(["Urban", "Rural"], num_rows)

# ----------------------------
# Lifestyle
# ----------------------------

sleep_hours = np.clip(np.random.normal(6.5, 1.5, num_rows), 4, 9)
stress_level = np.clip(np.random.normal(5, 2, num_rows), 1, 10)
social_media = np.clip(np.random.normal(3, 1.5, num_rows), 0.5, 8)

internet_access = np.random.choice(["Yes", "No"], num_rows)
participation_club = np.random.choice(["Yes", "No"], num_rows)
participation_sports = np.random.choice(["Yes", "No"], num_rows)
internship_completed = np.random.choice(["Yes", "No"], num_rows)
health_issues = np.random.choice(["Yes", "No"], num_rows)
part_time_job = np.random.choice(["Yes", "No"], num_rows)

# ----------------------------
# Final Score (Target)
# Strong Correlation Formula
# ----------------------------

final_exam_score = (
    attendance * 0.15 +
    internal1 * 0.2 +
    internal2 * 0.2 +
    assignment_avg * 0.15 +
    study_hours * 2 +
    previous_gpa * 5 -
    stress_level * 1.5 -
    backlogs * 5 +
    np.random.normal(0, 5, num_rows)
)

final_exam_score = np.clip(final_exam_score, 30, 100)

# Pass/Fail Classification
pass_fail = np.where(final_exam_score >= 50, "Pass", "Fail")

# ----------------------------
# Create DataFrame
# ----------------------------

data = pd.DataFrame({
    "attendance_percentage": attendance,
    "internal_exam_1": internal1,
    "internal_exam_2": internal2,
    "assignment_score_avg": assignment_avg,
    "quiz_avg": quiz_avg,
    "lab_performance": lab_perf,
    "previous_sem_gpa": previous_gpa,
    "backlog_count": backlogs,
    "project_score": project_score,
    "midterm_score": midterm_score,
    "study_hours_per_day": study_hours,
    "library_visits_per_month": library_visits,
    "online_course_count": online_courses,
    "doubt_sessions_attended": doubt_sessions,
    "age": age,
    "family_income": family_income,
    "distance_from_college_km": distance,
    "gender": gender,
    "parental_education_level": parent_education,
    "urban_or_rural": urban_rural,
    "sleep_hours": sleep_hours,
    "stress_level": stress_level,
    "social_media_hours": social_media,
    "internet_access_home": internet_access,
    "participation_club": participation_club,
    "participation_sports": participation_sports,
    "internship_completed": internship_completed,
    "health_issues": health_issues,
    "part_time_job": part_time_job,
    "final_exam_score": final_exam_score,
    "pass_fail": pass_fail
})

data.to_csv("student_performance_advanced.csv", index=False)

print("Dataset generated successfully with 3000 rows.")
print(data.head())
```

```python
import pandas as pd

data = pd.read_csv("student_performance_advanced.csv")
data.head()
```

```python
# Regression Target
y_reg = data["final_exam_score"]

# Classification Target
y_clf = data["pass_fail"]

# Drop targets from features
X = data.drop(columns=["final_exam_score", "pass_fail"])
```

```python
numerical_features = X.select_dtypes(include=["int64", "float64"]).columns
categorical_features = X.select_dtypes(include=["object"]).columns

print("Numerical:", len(numerical_features))
print("Categorical:", len(categorical_features))
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import train_test_split

numeric_pipeline = Pipeline([
    ("scaler", StandardScaler())
])

categorical_pipeline = Pipeline([
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer([
    ("num", numeric_pipeline, numerical_features),
    ("cat", categorical_pipeline, categorical_features)
])
```

```python
reg_pipeline = Pipeline([
    ("preprocessing", preprocessor),
    ("feature_selection", SelectKBest(score_func=f_regression, k=20)),
    ("model", RandomForestRegressor(random_state=42))
])

X_train, X_test, y_train, y_test = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

reg_pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

y_pred = reg_pipeline.predict(X_test)

print("R2 Score:", r2_score(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
clf_pipeline = Pipeline([
    ("preprocessing", preprocessor),
    ("model", RandomForestClassifier(random_state=42))
])

X_train, X_test, y_train, y_test = train_test_split(
    X, y_clf, test_size=0.2, random_state=42
)

clf_pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = clf_pipeline.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
import shap

# Fit only preprocessing
X_processed = reg_pipeline.named_steps["preprocessing"].fit_transform(X)

# Train a fresh RF for SHAP (needed for tree explainer)
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_processed, y_reg)

explainer = shap.TreeExplainer(rf_model)
shap_values = explainer.shap_values(X_processed[:200])  # limit for speed

shap.summary_plot(shap_values, X_processed[:200])
```

```python
# ==========================================
# 1️⃣ IMPORT LIBRARIES
# ==========================================

import pandas as pd
import numpy as np
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report


# ==========================================
# 2️⃣ LOAD DATASET
# ==========================================

df = pd.read_csv('/content/student_performance_advanced.csv')

print("Dataset Shape:", df.shape)


# ==========================================
# 3️⃣ DEFINE FEATURES
# ==========================================

X = df.drop(["final_exam_score", "pass_fail"], axis=1)

y_reg = df["final_exam_score"]
y_clf = df["pass_fail"]


# Separate numeric and categorical columns
numeric_features = X.select_dtypes(include=["int64", "float64"]).columns
categorical_features = X.select_dtypes(include=["object"]).columns


# ==========================================
# 4️⃣ PREPROCESSING PIPELINE
# ==========================================

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ]
)


# ==========================================
# 5️⃣ REGRESSION PIPELINE
# ==========================================

reg_pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("model", LinearRegression())
])

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

reg_pipeline.fit(X_train_reg, y_train_reg)

pred_reg = reg_pipeline.predict(X_test_reg)

print("\n🔹 REGRESSION RESULTS")
print("R2 Score:", r2_score(y_test_reg, pred_reg))
print("MSE:", mean_squared_error(y_test_reg, pred_reg))


# ==========================================
# 6️⃣ CLASSIFICATION PIPELINE
# ==========================================

clf_pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("model", LogisticRegression(max_iter=1000))
])

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X, y_clf, test_size=0.2, random_state=42
)

clf_pipeline.fit(X_train_clf, y_train_clf)

pred_clf = clf_pipeline.predict(X_test_clf)

print("\n🔹 CLASSIFICATION RESULTS")
print("Accuracy:", accuracy_score(y_test_clf, pred_clf))
print("\nClassification Report:\n")
print(classification_report(y_test_clf, pred_clf))


# ==========================================
# 7️⃣ SAVE MODELS (SMALL FILE SIZE)
# ==========================================

joblib.dump(reg_pipeline, "regression_model.pkl")
joblib.dump(clf_pipeline, "classification_model.pkl")

print("\n✅ Models saved successfully!")
```

```python
# ==========================================================
# 1️⃣ INSTALL SHAP (COLAB ONLY)
# ==========================================================
!pip install shap

# ==========================================================
# 2️⃣ IMPORT LIBRARIES
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import shap
import joblib

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score

plt.style.use("ggplot")

# ==========================================================
# 3️⃣ LOAD DATASET
# ==========================================================

df = pd.read_csv('/content/student_performance_advanced.csv')

# ==========================================================
# 4️⃣ VISUALIZATIONS (BEFORE ENCODING)
# ==========================================================

# Distribution of Final Exam Score
plt.figure(figsize=(8,5))
sns.histplot(df['final_exam_score'], bins=30, kde=True)
plt.title("Distribution of Final Exam Score")
plt.show()

# Correlation Heatmap (numerical only)
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# ==========================================================
# 5️⃣ ENCODE CATEGORICAL FEATURES
# ==========================================================

categorical_cols = [
    'gender',
    'parental_education_level',
    'urban_or_rural',
    'internet_access_home',
    'participation_club',
    'participation_sports',
    'internship_completed',
    'health_issues',
    'part_time_job'
]

label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Encode pass_fail
target_encoder = LabelEncoder()
df['pass_fail'] = target_encoder.fit_transform(df['pass_fail'])

# ==========================================================
# 6️⃣ FEATURES & TARGET
# ==========================================================

X = df.drop(columns=['pass_fail'])
y_class = df['pass_fail']
y_reg = df['final_exam_score']

# ==========================================================
# 7️⃣ TRAIN TEST SPLIT
# ==========================================================

X_train, X_test, y_train_class, y_test_class = train_test_split(
    X, y_class, test_size=0.2, random_state=42)

_, _, y_train_reg, y_test_reg = train_test_split(
    X, y_reg, test_size=0.2, random_state=42)

# ==========================================================
# 8️⃣ SCALING
# ==========================================================

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ==========================================================
# 9️⃣ MODEL TRAINING
# ==========================================================

clf = RandomForestClassifier(n_estimators=200, random_state=42)
reg = RandomForestRegressor(n_estimators=200, random_state=42)

clf.fit(X_train_scaled, y_train_class)
reg.fit(X_train_scaled, y_train_reg)

# ==========================================================
# 🔟 EVALUATION
# ==========================================================

class_pred = clf.predict(X_test_scaled)
reg_pred = reg.predict(X_test_scaled)

accuracy = accuracy_score(y_test_class, class_pred)
r2 = r2_score(y_test_reg, reg_pred)

print("🔥 Classification Accuracy:", accuracy)
print("🔥 Regression R2 Score:", r2)

# ==========================================================
# 1️⃣1️⃣ FEATURE IMPORTANCE (CLASSIFICATION)
# ==========================================================

plt.figure(figsize=(10,6))
importance = clf.feature_importances_
features = X.columns
sns.barplot(x=importance, y=features)
plt.title("Feature Importance - Classification")
plt.show()

# ==========================================================
# 1️⃣2️⃣ FEATURE IMPORTANCE (REGRESSION)
# ==========================================================

plt.figure(figsize=(10,6))
importance_reg = reg.feature_importances_
sns.barplot(x=importance_reg, y=features)
plt.title("Feature Importance - Regression")
plt.show()

# ==========================================================
# 1️⃣3️⃣ CROSS VALIDATION GRAPH
# ==========================================================

cv_scores = cross_val_score(clf, X_train_scaled, y_train_class, cv=5)

plt.figure(figsize=(6,4))
plt.plot(range(1,6), cv_scores, marker='o')
plt.title("Cross Validation Accuracy")
plt.xlabel("Fold")
plt.ylabel("Accuracy")
plt.show()

# ==========================================================
# 1️⃣4️⃣ MODEL COMPARISON
# ==========================================================

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train_scaled, y_train_class)

rf_acc = accuracy_score(y_test_class, clf.predict(X_test_scaled))
log_acc = accuracy_score(y_test_class, log_model.predict(X_test_scaled))

models = ['Random Forest', 'Logistic Regression']
accuracies = [rf_acc, log_acc]

plt.figure(figsize=(6,4))
plt.bar(models, accuracies)
plt.title("Model Comparison")
plt.ylabel("Accuracy")
plt.show()

# ==========================================================
# 1️⃣5️⃣ SHAP EXPLAINABILITY
# ==========================================================

explainer = shap.TreeExplainer(clf)
shap_values = explainer.shap_values(X_train_scaled)

shap.summary_plot(shap_values, X_train_scaled, feature_names=X.columns)

# ==========================================================
# 1️⃣6️⃣ SAVE MODELS (DEPLOYMENT SAFE)
# ==========================================================

joblib.dump(clf, "classification_model.pkl")
joblib.dump(reg, "regression_model.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(label_encoders, "categorical_encoders.pkl")
joblib.dump(target_encoder, "target_encoder.pkl")

print("✅ Everything saved successfully!")
```
